```
import numpy as np
from scipy import spatial
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def Glove(data):
  
  embeddings_dict = {}
  
  index = data.index
  columns = ["emb_A", "emb_B", "emb_P", "label"]
  emb = pd.DataFrame(index = index, columns = columns)
  emb.index.name = "ID"
  idx = 0

  with open("/content/drive/My Drive/GloVe_Wiki/glove.6B.200d.txt", 'r') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector

  for rowidx in range(0,len(data)):

    A = data.loc[rowidx,'A'].lower()
    print('A---------',A)
    B = data.loc[rowidx,'B'].lower()
    print('B---------',B)
    P = data.loc[rowidx,'Pronoun'].lower()
    print('P---------',P,'\n')

    if A not in embeddings_dict:
      continue
    A_embedding = np.array(embeddings_dict[A])

    if B not in embeddings_dict:
      continue
    B_embedding = np.array(embeddings_dict[B])

    if P not in embeddings_dict:
      continue
    P_embedding = np.array(embeddings_dict[P])

    label = "Neither"
    if (data.loc[rowidx,"A-coref"] == True):
      label = "A"
    if (data.loc[rowidx,"B-coref"] == True):
      label = "B"

    emb.iloc[idx] = [A_embedding,B_embedding,P_embedding,label]
    idx = idx + 1
  print('Index______________',idx)
  return (emb,idx)
```

```
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

!ls
# load data
dataset = pd.read_csv("/content/drive/My Drive/Dataset/Cleaned_Dataset_2.csv", sep = ',')

#data = data.sample(frac=1).reset_index(drop=True)
#train_data = data[:4930]
#print(len(train_data))
#test_data = data[4930:]
#print(len(test_data))

embedding_data, index = Glove(dataset)
#print(embedding_data[:index])

```

```
X = np.zeros((index,3*200))
Y = np.zeros((index,1))

print(type(embedding_data))
for i in range(0,index-1):
  A = np.array(embedding_data.loc[i,"emb_A"])
  B = np.array(embedding_data.loc[i,"emb_B"])
  P = np.array(embedding_data.loc[i,"emb_P"])

  X[i] = np.concatenate((A,B,P))
  
Y = embedding_data["label"][:index]

print(X)
print('_________________________________________________________')
print(Y)

seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

print(X_train)

# fit model no training data
model = XGBClassifier()

model.fit(X_train, y_train)

# make predictions for test data
y_pred = model.predict(X_test)

# evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
